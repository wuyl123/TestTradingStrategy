```python
import pandas as pd
from datetime import datetime
import numpy as np
from sklearn.linear_model import LinearRegression
```

按照$A = \beta*B + \alpha $形式构造数据，其中设$\beta$为2

```python
# 生成具有协整关系的价格序列  
np.random.seed(42)  
mins = 100000  

# 基础资产X  
x = np.cumprod(1 + np.random.normal(0.00005, 0.0002, mins)) * 100  

# 协整资产Y = 0.8X + 噪声  
spread = np.random.normal(0, 2, mins)  # 平稳价差  
y = 2.0 * x + spread  

A = pd.DataFrame(columns=['open','high','low','close','volume','datetime'])
B = pd.DataFrame(columns=['open','high','low','close','volume','datetime'])

A['close'] = x
B['close'] = y
```

```python
ref = pd.read_csv('ETHUSDT.csv')
A['datetime'] = ref['datetime'][:len(A)]
B['datetime'] = ref['datetime'][:len(B)]
```

```python
A['open'] = A['close']
A['high'] = A['close'] 
A['low'] = A['close'] 
B['open'] = B['close']
B['high'] = B['close'] 
B['low'] = B['close'] 
A['volume'] = A['close']
B['volume'] = B['close']
```

```python
# 验证拟合结果

model = LinearRegression()
x_B = B['close'].to_numpy()
x_A = A['close'].to_numpy()
model.fit(x_B[:3000].reshape(-1,1),x_A[:3000])
print(model.coef_[0],model.intercept_)
```
